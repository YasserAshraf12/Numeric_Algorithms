```python
# Gradient descent with learning rate = 0.01, x0=40, epsilon = 0.1
def first_order(f, x0, tol=0.1, n_iter=1000, alpha=0.01):
  for i in range(n_iter):
    
    x_new = x0 - alpha * Df(x0)
    x0 = x_new

    if abs(Df(x0)) <= tol:
      print("Found solution after ", (i+1), " iterations")
      print("r = ", x_new)
      print("Df(r1) = ", Df(x_new))
      break
```

```python
first_order(f, x0=40)
```

```python
# Newton's method without learning rate, x0=40, epsilon = 0.1
def newton_without_lr(f, x0, tol=0.1, n_iter=1000, alpha=0.01):
  for i in range(n_iter):
    x_new = x0 - (Df(x0) / DDf(x0))
    x0 = x_new
    
    if abs(Df(x0)) <= tol:
        print("Found solution after ", (i+1), " iterations")
        print("r = ", x_new)
        print("Df(r1) = ", Df(x_new))
        break
        
```

```python
newton_without_lr(f, x0=40)
```

```python
from numpy.linalg import inv, norm
```

```python
f = lambda x,y: x**2 + y**2
Df = lambda x, y: np.array([[2*x],
                            [2*y]])
```

```python
def newton_method_multi(x, h, alpha, epsilon, n_iter=1000):
    
    for i in range(n_iter):
        x = x - ((inv(h))@(alpha * Df(x[0][0], x[1][0])))
        
        if (norm(Df(x[0][0], x[1][0]))) <= epsilon:
            print("Found solution after ", (i+1), " iterations")
            print("r = ", list(x))
            print("Df(r1) = ", list(Df(x[0][0], x[1][0])))
            break
```

```python
h = [[2,0], [0, 2]]
```

```python
# First function
# Multi-variable Gradient descent with learning rate = 0.1,
# x0=(10,10), epsilon = 0.001
newton_method_multi(x=[[10], [10]], h=h, alpha=0.1, epsilon=0.001)
```

```python
# First function
# Multi-variable Newton's method without learning rate, 
# x0=(10,10), epsilon = 0.001
newton_method_multi(x=[[10], [10]], h=h, alpha=1, epsilon=0.001)
```

```python
# First function
# Multi-variable Newton's method with learning rate = 0.5, 
# x0=(10,10), epsilon = 0.001
newton_method_multi(x=[[10], [10]], h=h, alpha=0.5, epsilon=0.001)
```

```python
f = lambda x, y: (x**2) * y
Df = lambda x, y: np.array([[(2* x) * y], [x**2]])

h = lambda x, y: np.array([[(2 * y), (2*x)], [(2*x), 0]])
```

```python
def newton_method_multi_hesien(x, h, alpha, epsilon, n_iter=1000):
    for i in range(n_iter):
        
        he = h(x[0][0], x[1][0])
        
        x = x - ((inv(he))@(alpha * Df(x[0][0], x[1][0])))
        
        if (norm(Df(x[0][0], x[1][0]))) <= epsilon:
            print("Found solution after ", (i+1), " iterations")
            print("r = ", list(x))
            print("Df(r1) = ", list(Df(x[0][0], x[1][0])))
            break
```

```python
# Third function
# Multi-variable Gradient descent with learning rate = 0..01,
# x0=(10,10), epsilon = 0.001
newton_method_multi_hesien(x=[[10], [10]], h=h, alpha=0.1, epsilon=0.001)
```

```python
# Third function
# Multi-variable Newton's method without learning rate, 
# x0=(10,10), epsilon = 0.001
newton_method_multi_hesien(x=[[10], [10]], h=h, alpha=1, epsilon=0.001)
```
